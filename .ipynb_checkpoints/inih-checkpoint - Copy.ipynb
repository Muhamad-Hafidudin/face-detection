```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Mendefinisikan path ke folder 'train' dan 'test'
train_path = r"C:\Users\Hafidun\Machine Learning\Klasifikasi Gambar\Face Detection\Face\train"
test_path = r"C:\Users\Hafidun\Machine Learning\Klasifikasi Gambar\Face Detection\Face\test"
```

```python
# Mendefinisikan list untuk menyimpan gambar wajah dan labelnya
faces = []
labels = []

# Looping melalui setiap sub folder dalam folder 'train'
for folder_name in os.listdir(train_path):
    folder_path = os.path.join(train_path, folder_name)
    if os.path.isdir(folder_path):
        # Mendapatkan label dari nama sub folder
        label = folder_name

        # Looping melalui setiap file gambar dalam sub folder
        for image_name in os.listdir(folder_path):
            image_path = os.path.join(folder_path, image_name)
            # Membaca gambar menggunakan OpenCV
            image = cv2.imread(image_path)
            # Mengubah gambar menjadi grayscale
            gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            # Menambahkan gambar wajah dan label ke dalam list
            faces.append(gray_image)
            labels.append(label)

# Looping melalui setiap sub folder dalam folder 'test'
for folder_name in os.listdir(test_path):
    folder_path = os.path.join(test_path, folder_name)
    if os.path.isdir(folder_path):
        # Mendapatkan label dari nama sub folder
        label = folder_name

        # Looping melalui setiap file gambar dalam sub folder
        for image_name in os.listdir(folder_path):
            image_path = os.path.join(folder_path, image_name)
            # Membaca gambar menggunakan OpenCV
            image = cv2.imread(image_path)
            # Mengubah gambar menjadi grayscale
            gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            # Menambahkan gambar wajah dan label ke dalam list
            faces.append(gray_image)
            labels.append(label)
```

```python
# Mengubah list wajah dan label menjadi array numpy
faces = np.array(faces, dtype=np.uint8)
labels = np.array(labels)
```

```python
# Melakukan one-hot encoding terhadap label
le = LabelEncoder()
labels_encoded = le.fit_transform(labels)
```

```python
# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(faces, labels_encoded, test_size=0.2, random_state=42)
```

```python
# Menentukan ukuran dimensi yang diinginkan
desired_shape = (128, 128)

# Membagi data menjadi data latih dan data uji
X_train_resized = np.array([cv2.resize(face, desired_shape).flatten() for face in X_train])
X_test_resized = np.array([cv2.resize(face, desired_shape).flatten() for face in X_test])
```

```python
# Membuat objek SVC
svm = SVC(kernel='linear', probability=True)

# Melatih model menggunakan data latih
svm.fit(X_train_resized, y_train)
```

```python
# Memprediksi label untuk data uji
y_pred = svm.predict(X_test_resized)

# Mencetak laporan klasifikasi dan akurasi model
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Membaca gambar baru
new_image_path = r"C:\Users\Hafidun\Machine Learning\Klasifikasi Gambar\Face Detection\Face\jajall.png"
new_image = cv2.imread(new_image_path)

# Memeriksa apakah gambar berhasil dimuat
if new_image is not None:
    # Mengubah gambar menjadi skala abu-abu
    gray_new_image = cv2.cvtColor(new_image, cv2.COLOR_BGR2GRAY)

    # Melakukan prediksi kelas
    prediction = svm.predict(cv2.resize(gray_new_image, desired_shape).flatten().reshape(1, -1))

    # Mengubah label prediksi menjadi nama kelas
    predicted_class = le.inverse_transform(prediction)

    # Mencetak hasil prediksi
    print("Predicted Class:", predicted_class)
else:
    print("Failed to load image:", new_image_path)
    
# Visualisasi gambar baru
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
plt.title("Predicted Class: " + predicted_class[0])
plt.axis("off")
plt.show()
```
